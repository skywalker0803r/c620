# c620

```
f = F(config)
f.Recommended_mode = False
f.real_data_mode = False
f.c620_model.x_col
```

# load raw data

```
url = '/content/drive/MyDrive/台塑輕油案子/data/c620/raw/Aroma-2 C620+C660+C670 Plant Operation Matching Results_Actual Tray_0407.xlsx'
c620_feed = pd.read_excel(url,sheet_name='C620 Feed').T.reset_index(drop=True).T
c620_feed
```

# 計算c620_combined_feed_wt1

```
Feed_from_V615_ton = 175.068
Feed_from_V826_ton = 11.0939
V615_feed_wt = c620_feed.iloc[4:4+41,4]
V826_feed_wt = c620_feed.iloc[4:4+41,6]
total_ton = Feed_from_V615_ton + Feed_from_V826_ton 
Feed_from_V615_ton_p = Feed_from_V615_ton/total_ton 
Feed_from_V826_ton_p = Feed_from_V826_ton/total_ton
c620_combined_feed_wt = (V615_feed_wt*Feed_from_V615_ton_p)+(V826_feed_wt*Feed_from_V826_ton_p)
c620_combined_feed_wt = c620_combined_feed_wt.to_frame().T
c620_combined_feed_wt.columns = f.c620_model.x_col[-41:]
c620_combined_feed_wt
```

# 製作 c620_input dataframe1

```
c620_input1 = pd.DataFrame(index=[0],columns=f.c620_model.x_col)
c620_input1['Tatoray Stripper C620 Operation_Specifications_Spec 1 : Receiver Temp_oC'] = 38.2
c620_input1['Tatoray Stripper C620 Operation_Specifications_Spec 2 : Distillate Rate_m3/hr'] = 0.02
c620_input1['Tatoray Stripper C620 Operation_Specifications_Spec 3 : Benzene in Sidedraw_wt%'] = 80
c620_input1.update(c620_combined_feed_wt)
c620_input1
```

```
op_hat = f.c620_model.predict(c620_input1).iloc[:,41*4:]
op_hat
```

```
op_real = op_hat.copy()
op_real[:] = np.nan
op_real['Tatoray Stripper C620 Operation_Yield Summary_Reflux Rate_m3/hr'] = 138.96
op_real['Tatoray Stripper C620 Operation_Column Temp Profile_C620 Tray 14 (Control)_oC'] = 177.0
op_real['Tatoray Stripper C620 Operation_Column Temp Profile_C620 Tray 34 (Control)_oC'] = 194.7
op_real
```

```
op_col = op_real.dropna(axis=1).columns.tolist()
op_hat = op_hat[op_col]
op_real = op_real[op_col]
```

```
diff = op_hat - op_real
diff
```
