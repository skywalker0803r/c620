```python
from c620.F import F
from c620.config import config
import joblib
import numpy as np
import pandas as pd
import os
os.chdir('/content/c620')
print(os.getcwd())
!git pull
```

```python
f = F(config)
f.Recommended_mode = False
f.real_data_mode = False
f.c660_model.x_col
```

# load raw data

```python
url = '/content/drive/MyDrive/台塑輕油案子/data/c620/raw/Aroma-2 C620+C660+C670 Plant Operation Matching Results_Actual Tray_0407.xlsx'
c660_feed = pd.read_excel(url,sheet_name='C660 Feed').T.reset_index(drop=True).T
c660_feed
```

# 計算c660_combined_feed_wt1

```python
Feed_from_C620_Side_Draw_ton = 35.8488
Feed_from_T651_ton = 67.1833
C620_Side_Draw_wt = c660_feed.iloc[4:4+41,4]
T651_wt = c660_feed.iloc[4:4+41,6]
total_ton = Feed_from_C620_Side_Draw_ton + Feed_from_T651_ton
Feed_from_C620_Side_Draw_ton_p = Feed_from_C620_Side_Draw_ton/total_ton 
Feed_from_T651_ton_p = Feed_from_T651_ton/total_ton
c660_combined_feed_wt = (C620_Side_Draw_wt*Feed_from_C620_Side_Draw_ton_p)+(T651_wt*Feed_from_T651_ton_p)
c660_combined_feed_wt = c660_combined_feed_wt.to_frame().T
c660_combined_feed_wt.columns = f.c660_model.x_col[-41:]
c660_combined_feed_wt
```

# 製作 c660_input dataframe1

```python
c660_input1 = pd.DataFrame(index=[0],columns=f.c660_model.x_col)
c660_input1['Benzene Column C660 Operation_Specifications_Spec 2 : NA in Benzene_ppmw'] = 952
c660_input1['Benzene Column C660 Operation_Specifications_Spec 3 : Toluene in Benzene_ppmw'] = 3.7
c660_input1.update(c660_combined_feed_wt)
c660_input1
```

```python
op_hat = f.c660_model.predict(c660_input1).iloc[:,41*4:]
op_hat
```

```python
op_real = op_hat.copy()
op_real[:] = np.nan
op_real['Benzene Column C660 Operation_Yield Summary_Reflux Rate_m3/hr'] = 170.8947601
op_real['Benzene Column C660 Operation_Column Temp Profile_C660 Tray 6 (SD & Control)_oC'] = 88
op_real['Benzene Column C660 Operation_Column Temp Profile_C660 Tray 23 (Control)_oC'] = 91.74554443
op_real
```

```python
op_col = op_real.dropna(axis=1).columns.tolist()
op_hat = op_hat[op_col]
op_real = op_real[op_col]
```

```python
diff = op_hat - op_real
diff
```
